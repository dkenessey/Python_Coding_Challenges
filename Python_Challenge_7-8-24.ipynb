# Easy

Write a function that takes a string as input and returns the string reversed.

```python
input_string = "hello"
```

```python
def string_reversed(x):
  return x[::-1]
```

```python
string_reversed(input_string)
```

# Medium

Write a function that takes a sentence and returns the longest word in the sentence.

```python
input_sentence = "Data analysis is the key to success"
```

```python
def longest_word(sentence):
  words = sentence.split()
  return max(words, key=len)
```

```python
longest_word(input_sentence)
```

# Hard

Write a function that finds all anagrams in a list of words and groups them together.

```python
words = ["listen", "silent", "enlist", "hello", "ohell"]
```

```python
anagram_dict = {}

def anagrams(words):
    for word in words:
        sorted_word = ''.join(sorted(word))
        if sorted_word in anagram_dict:
            anagram_dict[sorted_word].append(word)
        else:
            anagram_dict[sorted_word] = [word]
    return list(anagram_dict.values())
```

```python
anagrams(words)
```
